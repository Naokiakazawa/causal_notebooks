## 因果推論

| カテゴリ                           | 詳細手法                                             | 説明 |
| ---------------------------------- | ---------------------------------------------------- | ------------------------------------------------------------------------------------------------------------------------------------------------------ |
| **ランダム化比較試験**             | 標準的なRCT（Randomized Controlled Trial）           | 対象を無作為に割り付け、交絡因子の影響を排除し真の因果効果を明確にする実験手法。                                                                      |
| **回帰分析による因果効果推定**     | OLS回帰                                            | 共変量を制御しながら線形関係を推定するシンプルな手法。基礎的な因果効果推定の出発点となる。                                                           |
|                                    | ロジスティック回帰                                  | バイナリアウトカムに対して対数オッズを推定し、処置効果を評価する手法。オッズ比で効果を解釈する。                                                      |
|                                    | 一般化線形モデル（GLM）                             | 異なる分布やリンク関数を利用し、非線形な関係や多様なアウトカム分布に対応可能な柔軟な回帰手法。                                                      |
| **傾向スコアを用いた手法**         | 傾向スコアマッチング                                | 各サンプルの処置受領確率（傾向スコア）を計算し、類似性のある処置群と対照群をマッチングして交絡因子を制御する手法。                                     |
|                                    | 傾向スコア重み付け（IPW）                            | 各サンプルに逆傾向スコアの重みを付与し、全体の平均処置効果（ATE）を推定する手法。極端なスコアに注意が必要。                                        |
| **Difference in Differences (DID)**| 標準的なDID手法                                     | 介入前後の差分を比較し、共通トレンド仮定の下で因果効果を推定するシンプルながら強力な手法。                                                            |
|                                    | シンセティックコントロール法                         | 複数の対照群から介入群に類似した合成対照群を構築し、介入効果をより正確に推定する手法。                                                               |
|                                    | CausalImpact                                       | ベイズ時系列モデルを用いて、時系列データ上で介入の影響を推定する手法。DID的アプローチの一例。                                                      |
|                                    | イベントスタディ分析                                | 複数のイベント時点での介入効果を時系列的に評価し、効果の時間変化を解析する手法。                                                                     |
| **回帰不連続デザイン（RDD）**      | シャープRDD                                        | 明確なカットオフに基づいて処置が決定される場合、その境界近傍のデータを用いて因果効果を推定する手法。                                                  |
|                                    | ファジーRDD                                        | カットオフ付近で処置割り当てが不完全な場合、処置確率の不連続性を利用して効果を推定する手法。                                                         |
|                                    | 局所線形回帰によるRDD                               | カットオフ近傍のデータに限定し、局所的に線形回帰を適用して処置効果を推定する手法。                                                                   |
|                                    | ノンパラメトリックRDD                               | パラメトリックな仮定を設けず、非線形な関係を柔軟に推定するための手法。                                                                               |
| **層化**                           | 完全層別 (Exact Stratification)                     | 共変量が完全に一致するサブグループを形成し、各層内で効果を推定することで交絡因子の影響を除去する方法。                                                |
|                                    | Coarsened Exact Matching (CEM)                      | 共変量を粗いカテゴリに分割して層を形成し、完全一致の厳密性とサンプル数のバランスを取る層化手法。                                                     |
|                                    | 傾向スコアに基づくサブクラス分け                     | 傾向スコアを利用してデータを複数のサブグループに分け、各層で効果を推定後に統合する手法。                                                           |
| **操作変数法**                     | 2段階最小二乗法 (2SLS)                               | 第一段階で操作変数から処置を予測し、第二段階でその予測値を用いてアウトカムを推定する標準的手法。                                                     |
|                                    | 限定情報最尤法 (LIML)                               | 2SLSの代替として、弱い器具変数下でより安定した推定結果を提供する手法。                                                                               |
|                                    | コントロールファンクションアプローチ               | 第一段階の残差を補正項としてアウトカムモデルに組み込むことで内生性問題を解決する手法。                                                              |
|                                    | 一般化モーメント法 (GMM)                             | 複数のモーメント条件を活用してパラメータを推定する、広範なモデルに適用可能な柔軟な手法。                                                             |
| **Meta-Learners**                  | T-Learner                                          | 処置群と対照群それぞれに個別のアウトカムモデルを学習し、その差分を効果として算出する手法。                                                           |
|                                    | S-Learner                                          | 全データを一つのモデルで学習し、処置変数を含めた予測から直接効果を推定するシンプルなアプローチ。                                                     |
|                                    | X-Learner                                          | 不均衡なサンプルに対して、異質な処置効果を補正し効率的に効果を推定する手法。                                                                         |
|                                    | DR-Learner (Doubly Robust Learner)                  | 傾向スコアとアウトカムモデルの両方を組み合わせ、どちらか一方が誤指定されてもロバストな効果推定を実現する手法。                                        |
| **因果フォレスト**                 | 標準的なCausal Forest                              | ランダムフォレストの概念を拡張し、非パラメトリックに各サンプルの処置効果（CATE）を推定する手法。                                                     |
|                                    | 一般化ランダムフォレストによるCATE推定              | ランダムフォレストアルゴリズムを用い、サンプルごとに柔軟な処置効果を推定する手法。                                                                 |
|                                    | ベイズ的因果フォレスト (BART等)                     | ベイズ的手法を取り入れ、分布全体の不確実性を考慮した非パラメトリックな因果効果推定を行う手法。                                                       |
| **感度分析**                       | Rosenbaum Bounds                                   | 未観測交絡の影響の大きさを数値化し、推定結果の頑健性を評価する手法。                                                                                  |
|                                    | プレースボテスト                                   | 処置変数を無関係な変数に置換して因果効果が消失するか検証することで、推定の信頼性を評価する手法。                                                      |
|                                    | ランダム共通原因変数の追加                           | 追加の共変量を導入し、推定結果の変動を確認することで、未観測交絡の影響を評価する方法。                                                               |
|                                    | サブセット分析                                     | データの一部を除外して推定を再実施し、結果の一貫性と頑健性を検証する手法。                                                                          |
|                                    | E-value計算                                        | 未観測交絡があった場合に、推定効果を相殺するために必要な交絡の強さを定量化する指標。                                                               |
| **ベイズ推定**                     | MCMC: Metropolis-Hastings                          | 複雑な事後分布からサンプルを生成する基本的なMCMCアルゴリズム。幅広く利用され、シンプルな実装が可能。                                                  |
|                                    | MCMC: Gibbs Sampling                               | 条件付き分布に基づき反復的にサンプルを生成する手法。連続変数の推定などに適用される。                                                                |
|                                    | MCMC: Hamiltonian Monte Carlo (HMC)/NUTS           | 勾配情報を利用して効率的に高次元分布からサンプルを生成する先進的なMCMC手法。                                                                         |
|                                    | Variational Inference: Mean-field                  | 複雑な事後分布を単純な独立分布で近似し、最適化手法を用いてパラメータ推定を行う手法。                                                                 |
|                                    | Variational Inference: Black-box                   | 自動微分を利用して、汎用的に複雑なベイズモデルの近似推定を実施する手法。                                                                             |
|                                    | Bayesianモデリング: Bayesian Linear Regression     | 線形回帰モデルをベイズ的に推定し、パラメータの事後分布を得ることで不確実性を評価する手法。                                                            |
|                                    | Bayesianモデリング: Bayesian Hierarchical Models   | 階層構造を持つデータに対してグループ間の変動を同時に推定する柔軟なベイズモデル。                                                                     |
|                                    | Bayesianモデリング: BSEM                           | 構造方程式モデルをベイズ枠組みで推定し、複雑な因果構造を同時に解析する手法。                                                                         |
